### 3.3 Preprocessing

- Dealing with missing values<br/>
    - Dropping the column __'keys'__ because of it __100%__ uniques.
    - Dropping the column __'country'__ which holds the constant value __'US'__
    - Dropping the column __'websites'__ as it has __465__ values and not required for my analysis as well.

```python
res_df['keys'].nunique()   #Columns key holds the unique key num of restaurants
```

```python
res_df = res_df.drop(['keys', 'websites','country'], axis=1) #Dropping keys and websites column 

#rep_df = res_df #assigning the dataframe to new variable before adding a extra column 'Count'



res_df.head()
```

```python
res_df['Count'] = res_df.groupby(['province'])[['name']].transform('count'); 
```

Adding a new column __'Count'__ by grouping the restaurants based on postalcode to get to now which are has the leading restaurants countS across the region.

We have messy data in the column name so i have decided to clean it up by capitalizing the strings and removing the special characters by using the below code.  

```python
res_df.name = res_df.name.apply(lambda x: x.upper())
res_df['name'] = res_df['name'].str.replace(" ","")
res_df['name'] = res_df['name'].str.replace('[^\w\s]','')
```

Created new dataframe __group_res__ by grouping the data to get the count of restaurants located in each city of each province.

```python
group_res =res_df.groupby(['name'], as_index = True)['Count'].count()
```

```python
group_res.nlargest(10)
```

After cleaning the data in column __name__ the restaurant count has increased for ex: __MCD__ had __1886__ in count across the country now it has raised to __2116__. Because the original dataset had same restaurants names in lower letter with additional special character.

```python
rep_df = res_df.drop(['Count'], axis=1) 

#rep_df = res_df #assigning the dataframe to new variable before adding a extra column 'Count'
```

#### 4.9 Spread of Mcdonalds across the top 5 states

```python
mcd_subset = res_df[(res_df.province == 'CA') | (res_df.province == 'OH')| (res_df.province == 'TX')| (res_df.province == 'FL') | (res_df.province == 'IN')]
```

```python
mcd_id_subset = mcd_subset.copy()
mcd_id_subset['mcd_id'] = mcd_id_subset['name'].apply(lambda i: 1 if i == 'MCDONALDS' 
                                                      else 0)

mcd_id_subset['mcd_id'].unique()
```

```python
lm = sns.lmplot(x="latitude", y="longitude", data=mcd_id_subset, hue = 'mcd_id', fit_reg = False,
          legend = True, markers=["o", "x"])
axes = lm.axes
axes[0, 0].set_title('Coords of Mcdonalds across CA,OH,TX,FL,IN - (Gold Star in Orange)')
axes[0,0].set_ylim(-140, -70)
```

#### 4.11Spread of Restaurants across US(Map View)

```python
res_df['text'] = res_df['name'] + ',' + res_df['province'] + ', ' + res_df['country']

scl = [ [0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
    [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"] ]

data = [ dict(
        type = 'scattergeo',
        locationmode = 'USA-states',
        lon = res_df['longitude'],
        lat = res_df['latitude'],
        text = res_df['text'],
        mode = 'markers',
        marker = dict(
            size = 4,
            opacity = 0.5,
            reversescale = True,
            autocolorscale = False,
            symbol = 'circle',
            line = dict(
                width=.5,
                color='rgba(102, 102, 102)'
            )))]
layout = dict(
        title = 'Restaurants across the country',
        colorbar = True,
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showland = True,
            landcolor = "rgb(250, 250, 250)",
            subunitcolor = "rgb(217, 217, 217)",
            countrycolor = "rgb(217, 217, 217)",
            countrywidth = 0.5,
            subunitwidth = 0.5
        ),
    )

fig = dict( data=data, layout=layout )
iplot( fig, validate=False, filename='all restaurants' )
```
